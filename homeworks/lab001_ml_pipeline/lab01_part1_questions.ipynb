## ex. 4

Classic matrix factorization example. Given matrix $X$ you need to find $A$, $S$ to approximate $X$. This can be done by simple gradient descent iteratively alternating $A$ and $S$ updates.
$$
J = || X - AS ||_F^2  , \quad A\in \mathbb{R}^{N \times R} , \quad S\in \mathbb{R}^{R \times M}
$$
$$
\frac{dJ}{dS} = ?
$$

You may use one of the following approaches:

#### First approach
Using ex.2 and the fact:
$$
|| X ||_F^2 = tr(XX^T)
$$

$$
J=||X-AS||_F^2=tr((X-AS)^T(X-AS))=tr(X^TX-X^TAS-S^TA^TX+S^TA^TAS)
$$
$$
\frac{dJ}{dS}=-A^TX-AX^T+2A^TAS=2A^T(AS-X)
$$

it is easy to derive gradients (you can find it in one of the refs).

#### Second approach
You can use *slightly different techniques* if they suits you. Take a look at this derivation:
<img src="https://github.com/girafe-ai/ml-course/blob/26s_harbour/homeworks/lab001_ml_pipeline/grad.png?raw=1">
(excerpt from [Handbook of blind source separation, Jutten, page 517](https://books.google.ru/books?id=PTbj03bYH6kC&printsec=frontcover&dq=Handbook+of+Blind+Source+Separation&hl=en&sa=X&ved=0ahUKEwi-q_apiJDLAhULvXIKHVXJDWcQ6AEIHDAA#v=onepage&q=Handbook%20of%20Blind%20Source%20Separation&f=false), open for better picture).

From the Euclidean distance cost function:
$
J_E = D_E(\mathbf{X}; AS) = \frac{1}{2}\|X - AS\|_F^2 = \frac{1}{2}\sum_{pt}(x_{pt} - [AS]_{pt})^2
$

Taking the partial derivative with respect to $s_{nt}$:
$
\frac{\partial J_E}{\partial s_{nt}} = -[A^T X]_{nt} + [A^T AS]_{nt} = -([A^T X]_{nt} - [A^T AS]_{nt})
$

$
\frac{\partial J}{\partial S} = A^T(AS - X)
$

#### Third approach
And finally we can use chain rule!
let $ F = AS $

**Find**
$$
\frac{dJ}{dF} =  2(AS-X)^T
$$
and
$$
\frac{dF}{dS} = A \otimes I_M
$$
(the shape should be $ NM \times RM )$.

Now it is easy do get desired gradients:
$$
\frac{dJ}{dS} =  \frac{\partial J}{\partial S} = 2A^T(AS - X)
$$

### Question 2

We can also use other distance metrics such as L1 distance.
For pixel values $p_{ij}^{(k)}$ at location $(i,j)$ of some image $I_k$,

the mean $\mu$ across all pixels over all images is $$\mu=\frac{1}{nhw}\sum_{k=1}^n\sum_{i=1}^{h}\sum_{j=1}^{w}p_{ij}^{(k)}$$
And the pixel-wise mean $\mu_{ij}$ across all images is
$$\mu_{ij}=\frac{1}{n}\sum_{k=1}^np_{ij}^{(k)}.$$
The general standard deviation $\sigma$ and pixel-wise standard deviation $\sigma_{ij}$ is defined similarly.

Which of the following preprocessing steps will not change the performance of a Nearest Neighbor classifier that uses L1 distance? Select all that apply.
1. Subtracting the mean $\mu$ ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu$.)
2. Subtracting the per pixel mean $\mu_{ij}$  ($\tilde{p}_{ij}^{(k)}=p_{ij}^{(k)}-\mu_{ij}$.)
3. Subtracting the mean $\mu$ and dividing by the standard deviation $\sigma$.
4. Subtracting the pixel-wise mean $\mu_{ij}$ and dividing by the pixel-wise standard deviation $\sigma_{ij}$.
5. Rotating the coordinate axes of the data.

*Your Answer:* Options 1 , 2 and 3


*Your Explanation:*

1. Subtracting the mean $\mu$This is just shifting the data.

When you calculate the distance between image $a$ and image $b$, the $\mu$ just cancels out:
$$|(p_{ij}^{(a)} - \mu) - (p_{ij}^{(b)} - \mu)| = |p_{ij}^{(a)} - p_{ij}^{(b)}|$$
The distance is identical, so the result is identical.

2. Subtracting the per pixel mean $\mu_{ij}$

Even though $\mu_{ij}$ is different for every pixel, it is the same value for both images at that location. It cancels out during the subtraction:
$$|(p_{ij}^{(a)} - \mu_{ij}) - (p_{ij}^{(b)} - \mu_{ij})| = |p_{ij}^{(a)} - p_{ij}^{(b)}|$$
Distance is identical again.

3. Subtracting $\mu$ and dividing by $\sigma$

The subtraction part cancels out like in step 1. The division by $\sigma$ just scales all distances by the exact same number:
($\frac{1}{\sigma}$).$$\frac{1}{\sigma} |p_{ij}^{(a)} - p_{ij}^{(b)}|$$
Since every distance shrinks by the same ratio, the "closest" neighbor is still the closest.



## Question 3

Which of the following statements about $k$-Nearest Neighbor ($k$-NN) are true in a classification setting, and for all $k$? Select all that apply.
1. The decision boundary (hyperplane between classes in feature space) of the k-NN classifier is linear.
2. The training error of a 1-NN will always be lower than that of 5-NN.
3. The test error of a 1-NN will always be lower than that of a 5-NN.
4. The time needed to classify a test example with the k-NN classifier grows with the size of the training set.
5. None of the above.

*Your Answer:*

4) The time needed to classify a test example with the k-NN classifier grows with the size of the training set.


*Your Explanation:*

1) FALSE  The decision boundary definitely isn't a simple straight line. Since $k$-NN just looks at the closest neighbors, the boundary ends up being this complex, wiggly shape that wraps around specific points. It’s locally simple, but globally, it’s definitely non-linear.
2) FALSE  This is false because of the word "always." Sure, 1-NN usually has 0 training error because it just memorizes the data. But 5-NN can also have 0 error if the groups are clearly separated. Since they can be equal (both 0), saying 1-NN is strictly "lower" isn't right.
3) FALSE  Actually, it's usually the opposite. 1-NN overfits really badly, it picks up on all the noise and outliers. 5-NN is smoother because it averages things out, so it usually works better on new test data (lower error), not worse.
4) TRUE  $k$-NN is a "lazy" algorithm, it doesn't really learn a model beforehand. When you want to classify a new point, it has to compare that point to every single training image to find the distance. So, if you double the number of training images ($N$), you have to do double the work. That means the time grows linearly, or $O(N)$.
